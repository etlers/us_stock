```python
import pandas as pd

# 연속 로그 출력
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import warnings
warnings.filterwarnings("ignore")
```

```python
import yfinance as yf

msft = yf.Ticker("MSFT")
```

```python
# get stock info
msft.info
```

```python
# get historical market data
hist = msft.history(period="max")
hist.reset_index()
```

```python
# show actions (dividends, splits)
msft.actions
```

```python
# show dividends
msft.dividends
```

```python
# show splits
msft.splits
```

```python
# show financials
msft.financials
msft.quarterly_financials
```

```python
# show major holders
msft.major_holders
```

```python
# show institutional holders
msft.institutional_holders
```

```python
# show balance sheet
msft.balance_sheet
msft.quarterly_balance_sheet
```

```python
# show cashflow
msft.cashflow
msft.quarterly_cashflow
```

```python
# show earnings
msft.earnings.reset_index()
msft.quarterly_earnings.reset_index()
```

```python
# show sustainability
msft.sustainability.reset_index()
```

```python
# show analysts recommendations
msft.recommendations.reset_index()
```

```python
# show next event (earnings, etc)
msft.calendar
```

```python
# show all earnings dates
msft.earnings_dates.reset_index()
```
