```python
def find_end_vs_low_price_rate(dict_ticker):
    low_prcie = dict_ticker['52 Week Range'].split('-')[0].strip()
    high_prcie = dict_ticker['52 Week Range'].split('-')[1].strip()
    end_prcie = dict_ticker['Previous Close']

    price_gap = end_prcie - float(low_prcie.replace(',',''))
    vs_rate = round(price_gap / float(low_prcie.replace(',','')) * 100, 2)

    return vs_rate, low_prcie, high_prcie, end_prcie
```
